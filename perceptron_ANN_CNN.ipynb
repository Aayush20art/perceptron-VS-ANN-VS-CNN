```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tensorflow.keras import Sequential #it lets you build a neural network layer by layer in Keras
from tensorflow.keras.layers import Dense #Dense makes the final prediction
from tensorflow.keras.layers import Convolution2D #it extracts features
from tensorflow.keras.layers import Flatten #it convers all into 1-D
from tensorflow.keras.layers import MaxPooling2D #it reduces the size
from tensorflow.keras.layers import Dropout #Dropout prevents overfitting
from tensorflow.keras.utils import to_categorical #convert numeric class labels into one hot encoded format for training c;assification models

```

```python
x = df.drop('label', axis = 1)
y = df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
```

```python
x_train_img = x_train.values.reshape(-1, 28, 28)
x_test_img = x_test.values.reshape(-1, 28, 28)
```

```python
x_train_img
```

```python
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)
```

```python
#perceptron model
perceptron = Sequential([
    Flatten(input_shape = (28,28)),
    Dense(10, activation = "softmax")
])
```

```python
perceptron.compile(optimizer= 'sgd', loss = 'categorical_crossentropy', metrics = ["accuracy"])
```

```python
history_percp = perceptron.fit(x_train_img, y_train_cat, epochs = 5,
                               batch_size = 32, validation_data = (x_test_img, y_test_cat), verbose = True)
```

```python
acc_percp = perceptron.evaluate(x_test_img, y_test_cat, verbose = 0)[1]
```

```python
acc_percp
```

```python
#ANN
ann = Sequential([
    Flatten(input_shape = (28,28)),
    Dense(128, activation = "relu"),
    Dense(64, activation = "relu"),
    Dense(10, activation = "softmax")
])
```

```python
ann.compile(optimizer= 'adam', loss= 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
history_ann = ann.fit(x_train_img, y_train_cat, epochs = 5, batch_size=32, validation_data=(x_test_img, y_test_cat), verbose = True)
```

```python
acc_ann = ann.evaluate(x_test_img, y_test_cat, verbose = 0)[1]
```

```python
acc_ann
```

```python
x_train_cnn = x_train.values.reshape(-1,28,28,1) #if we have to use RGB image then chnage 1 to 3
x_test_cnn = x_test.values.reshape(-1,28,28,1)
```

```python
#CNN
cnn = Sequential([
    Convolution2D(32, kernel_size=(3,3), activation = "relu", input_shape = (28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    Convolution2D(64, kernel_size = (3,3),activation= "relu"),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(128,activation="relu"),
    Dropout(0.5),
    Dense(10, activation="softmax")
])
```

```python
cnn.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics = ["accuracy"])
```

```python
history_cnn = cnn.fit(x_train_cnn, y_train_cat, epochs = 5, batch_size= 32, validation_data= (x_test_cnn, y_test_cat), verbose = True)
```

```python
acc_cnn = cnn.evaluate(x_test_cnn, y_test_cat, verbose = 0)[1]
```
